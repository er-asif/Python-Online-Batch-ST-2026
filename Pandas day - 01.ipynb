# Pandas
### Pandas is a popular python library used for data analysis and manipulation.
### It provides two main data structures :-
* Series - 1 Dimensional labeled array
* Dataframes - 2 Dimensional table. 
  

```python
import pandas as pd
```

```python
print(pd.__version__)
```

## Series

```python
s = pd.Series([10,20,30,40,50,60])
print(s)
```

### Custom index

```python
s = pd.Series([10,20,30,40,50,60], index=["A","B", "C", "D", "E", "F"])
print(s)
```

### Access 


```python
print(s["B"])
```

## DataFrame

```python
data = {
    "name" : ["Asif", "Aman", "Alice", "Charlie"],
    "age" : [23, 22, 18, 34],
    "city" : ["Lucknow", "Guwahti", "London", "New York"] 
}
df = pd.DataFrame(data)
print(df)
```

## Info

```python
df.info()
```
