# Reshape method
### Used to reshape numpy array

```python
arr = np.array([
    [8,9,10],
    [11,12,13]
])

print("Shape of ARR =", arr.shape)

new_arr = arr.reshape(3,2)
print(new_arr)
print("Shape of New array", new_arr.shape)
```
